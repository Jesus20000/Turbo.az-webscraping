```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re 
```

```python
turboaz_data = pd.DataFrame()
```

```python
for page_number in range(1,5):
    url = f'https://turbo.az/autos?page={page_number}'

    page = requests.get(url)

    # convert to html
    html = BeautifulSoup(page.content, 'html.parser')

    carinfo = html.find_all('div', class_ = 'products-i__bottom')

    for index in range(len(carinfo)):
        try:
            cardetails = re.sub('<[^>]+>', '/', str(carinfo[index]))

            price = cardetails.split('/')[3]
            currency = cardetails.split('/')[4]
            make_model = cardetails.split('/')[8]
            others = cardetails.split('/')[10]
            year = others.split(',')[0].strip()
            engine = others.split(',')[1].strip()
            km = others.split(',')[2].strip()


            turboaz_data = turboaz_data.append(
            {
                'Make Model':make_model,
                'Price':price,
                'Currency':currency,
                'Production year':year,
                'Engine Power':engine,
                'KM':km
            }, ignore_index=True)
        except:
            continue
```
